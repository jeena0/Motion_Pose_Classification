# Training model

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
x_train_std=std.fit_transform(x_train)
x_test_std=std.transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier() #algorithm - Decision Tree
```

```python
dt.fit(x_train_std,y_train)
```

```python
dt.score(x_train_std,y_train)
```

```python
y_pred=dt.predict(x_test_std)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy*100) 
# the phenomenon most likely due to over-fitting
```

```python
x_train.duplicated().sum() 
```

```python
#No dulpicate rows found.
```

```python
#cross validation
from sklearn.model_selection import cross_val_score
dt_cv_scores= cross_val_score(dt, x_train_std,y_train, cv=5)
dt_mean=dt_cv_scores.mean()
print(f'cross validation score {dt_cv_scores}')
print(f'Mean CV Score is {dt_mean}')
```

this is not a case of over-fitting since by evaluating cross validation in decision tree, we got a score near 100%

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train_std,y_train['pose'])
```

```python
lr.score(x_train_std,y_train)
```

```python
pred_lr=lr.predict(x_test_std)
```

```python
accuracy = accuracy_score(y_test, pred_lr)
print("Accuracy Score:", accuracy*100)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train_std,y_train["pose"])
```

```python
knn.score(x_train_std,y_train)
```

```python
pred_knn= knn.predict(x_test_std)
```

```python
accuracy=accuracy_score(y_test, pred_knn)
print("Accuracy :",accuracy*100)
```
